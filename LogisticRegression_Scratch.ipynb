```python
def sigmoid(x):
    res = 1/(1+np.exp(-x))
    return res
```

```python
#from sklearn.model_selection import train_test_split
#train, test = train_test_split(df1, test_size=0.2)
score1 = df1[["exam1","exam2"]]

output = df1['y']
ones = np.ones([score1.shape[0], 1]) 
score1 = np.concatenate([ones, score1], 1)
weights = np.zeros(score1.shape[1])
```

```python
def cost(theta, x, y):
    m = x.shape[0]
    return - (1 / m) * np.sum(y * np.log(sigmoid(product(theta,x))) + (1 - y) * np.log(1 - sigmoid(product(theta,x))))
```

```python
def product(weights,x):
    return np.dot(x, weights)
```

```python
def diff(weights,x,y):
    pred = []
    for i in y:
        pred.append(i)
        m = x.shape[0]
    return (1 / m) * np.dot(x.T, sigmoid(product(weights, x)) - pred)

def LogisticRegression(X,Y,lr,iterations,weights):
    m = X.shape[0]
    limit = 0.25
    costval = 10
    costlist = []
    costlist.append(costval)
    iterations = 0
    while costval > limit:                       
        costval = cost(weights,X,Y)                   
        diffval = diff(weights,X,Y)                  
        weights = weights - (lr*diffval)                     
        if iterations%100==0:                  
            print(cost)
            costlist.append(costval)
        iterations = iterations + 1                  
      

    return weights, costval, costlist
```
